# ベクトル
ベクトルとベクトル空間は、*線形代数*の基本であり、多くの機械学習モデルで使われています。ベクトルは空間内の線や平面を記述し、多次元空間の関係を探る計算を可能にします。

## ベクトルとは
最も単純な形では、ベクトルは*大きさ*と*方向*の両方を持つ数値要素です。大きさは距離（例えば「2マイル」）を表し、方向はベクトルの向かう方向（例えば「東」）を示します。ベクトルは、任意の原点から線で結ぶことができる空間内の点を記述するn次元座標によって定義されます。

それは少し複雑に聞こえるかもしれませんが、簡単な2次元の例から始めましょう。この場合、2次元平面内の点によって定義されるベクトルを使用します。2次元座標は*x*と*y*の値で構成され、この場合は*x*に**2**、*y*に**1**を使用します。

私たちのベクトルは**v**=(2,1)と書けますが、より正式には、ベクトルの次元座標値が行列として示される次の表記を使用します。
$$\vec{v} = \begin{bmatrix}2 \\ 1 \end{bmatrix}$$

では、これは一体何を意味するのでしょうか？座標は2次元であり、ベクトルの終点（または*頭*）に到達するために必要な動きを記述します。この場合、*x*次元に2単位、*y*次元に1単位移動する必要があります。ベクトルの開始点は指定しないことに注意してください。単にベクトルの大きさ（マグニチュード）と方向をカプセル化する目的地の座標を記述しているだけです。*ここ*がどこであるかを指定せずに、*そこ*に到達するために従う必要がある指示と考えてください！

ベクトルを視覚化すると役立ちます。2次元ベクトルでは、それは非常に簡単です。単に2次元平面を定義し、開始点を選択し、開始点に対してベクトルによって記述された座標をプロットするだけです。

次のセル内のコードを実行して、ベクトル**v**（座標(2,1)で記述されます）を視覚化します。

````python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

# We'll use a numpy array for our vector
v = np.array([2,1])

# and we'll use a quiver plot to visualize it.
origin = [0], [0]
plt.axis('equal')
plt.grid()
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.quiver(*origin, *v, scale=10, color='r')
plt.show()
````

Pythonでベクトルを定義するには、numpy配列を使用できることに注意してください。したがって、(2,1)ベクトルを作成するには、単純に要素[2,1]を持つnumpy配列を作成します。次に、クィバープロットを使用してベクトルを視覚化しました。0,0の点を開始点（または*原点*）として使用しています。(2,1)のベクトルは、0,0から始まり、*x*軸に沿って2単位（右に）、*y*軸に沿って1単位（上に）移動する矢印として示されています。

## ベクトルの大きさ（マグニチュード）と方向の計算
私たちは、ベクトルの成分を*デカルト座標*で表すことによって、ベクトルを扱う傾向があります。つまり、各次元に沿って移動する単位数を定義する*x*と*y*（およびその他の次元）の値です。したがって、(2,1)ベクトルの座標は、*x*軸に沿って2単位、*y*軸に沿って1単位移動する必要があることを示しています。

ただし、ベクトルの*極座標*、つまりベクトルの大きさ（マグニチュード）と方向を記述する座標でベクトルを扱うこともできます。大きさは、ベクトルの尾から頭までの全体の距離であり、方向はベクトルが向いている角度です。

### 大きさ（マグニチュード）の計算
ベクトルのデカルト座標から大きさ（マグニチュード）を計算するには、任意の開始点とベクトルの頭の点との間の距離を測定する必要があります。2次元ベクトルでは、実際には直角三角形の斜辺の長さを計算しているだけなので、ピタゴラスの定理を呼び出して、その成分の二乗の合計の平方根を計算することができます。次のようになります。

$$\|\vec{v}\| = \sqrt{v_{1}\;^{2} + v_{2}\;^{2}}$$

ベクトルの大きさの表記は、ベクトル名を縦線で囲むことです。単一の線（例：|**v**|）または二重の線（例：||**v**||）を使用できます。二重の線は、絶対値との混同を避けるためによく使用されます。ベクトルの成分は、添字（v<sub>1</sub>, v<sub>2</sub>,...v<sub>*n*</sub>）で示されることに注意してください。

この場合、ベクトル**v**は値**2**と**1**の2つの成分を持つため、大きさの計算は次のようになります。

$$\|\vec{v}\| = \sqrt{2^{2} + 1^{2}}$$

これは次のようになります。

$$\|\vec{v}\| = \sqrt{4 + 1}$$

したがって、次のようになります。

$$\|\vec{v}\| = \sqrt{5} \approx 2.24$$

以下のPythonコードを実行して、より正確な結果を得ることができます（numpy配列の要素はゼロベースであることに注意してください）。

````python
import math

vMag = math.sqrt(v[0]**2 + v[1]**2)
print (vMag)
````

この計算は任意の次元のベクトルに機能します。単に、二乗された成分の合計の平方根を取るだけです。

$$\|\vec{v}\| = \sqrt{v_{1}\;^{2} + v_{2}\;^{2} ... + v_{n}\;^{2}}$$

Pythonでは、*numpy*は、ベクトルをより簡単に扱うための線形代数ライブラリである**linalg**を提供しています。以下のコードの**norm**関数を使用して、ベクトルの大きさを計算できます。

````python
import numpy as np

vMag = np.linalg.norm(v)
print (vMag)
````

### 方向の計算
ベクトルのデカルト座標から方向、または*振幅*を計算するには、少し三角法を使用する必要があります。*逆正接*（*arctan*とも呼ばれます）を計算することで、ベクトルの角度を得ることができます（*正接*は角度を比率で計算します。逆正接、または**tan<sup>-1</sup>**はこれを度で表します）。

任意の直角三角形では、正接は*対辺*を*隣辺*で割って計算されます。2次元ベクトルでは、これは*y*値を*x*値で割ったものであり、**v**ベクトル(2,1)の場合、次のようになります。

$$\tan(\theta) = \frac{1}{2}$$

これにより結果***0.5***が生成され、そこから電卓を使用して逆正接を計算し、角度を度で得ることができます。

$$\theta = \tan^{-1} (0.5) \approx 26.57^{o}$$

方向の角度は***&theta;***として示されることに注意してください。

以下のPythonコードを実行して、これを確認してください。

````python
import math
import numpy as np

v = np.array([2,1])
vTan = v[1] / v[0]
print ('tan = ' + str(vTan))
vAtan = math.atan(vTan)
# atan returns the angle in radians, so convert to degrees
print('inverse-tan = ' + str(math.degrees(vAtan)))
````

しかし、追加の複雑さがあります。*x*または*y*（またはその両方）の値が負の場合、ベクトルの向きが標準ではなく、電卓が間違ったtan<sup>-1</sup>値を返す可能性があります。ベクトルの正しい方向を確実に得るには、次のルールを使用します。
- *x*と*y*の両方が正の場合：tan<sup>-1</sup>値を使用します。
- *x*が負で、*y*が正の場合：tan<sup>-1</sup>値に180を加算します。
- *x*と*y*の両方が負の場合：tan<sup>-1</sup>値に180を加算します。
- *x*が正で、*y*が負の場合：tan<sup>-1</sup>値に360を加算します。

なぜこれを行う必要があるのか理解するために、このように考えてみましょう。ベクトルは360度の円弧の任意の方向を指すことができます。その円を、中央にx軸とy軸を持つ4つの象限に分割します。角度は、x軸から正（反時計回り）と負（時計回り）の両方向で測定できます。象限は正（反時計回り）の方向（正の角度を測定する方法）で次のように番号付けします。

```
    2 | 1
    - o -
    3 | 4
```

では、4つのベクトルの例を見てみましょう。

1. ベクトル[2,4]は、xとyの両方に正の値を持っています。このベクトルの線は、象限3から象限1へ0,0の点を通ります。4/2のTan<sup>-1</sup>は約63.4度で、これはx軸からベクトル線までの正の角度です。したがって、これがベクトルの方向です。
2. ベクトル[-2,4]は、負のxと正のyを持っています。このベクトルの線は、象限4から象限2へ0,0の点を通ります。4/-2のTan<sup>-1</sup>は約-64.4度で、これはx軸からベクトル線までの*負の*角度ですが、方向が間違っています（ベクトルが象限2から象限4に向かって移動しているかのように）。したがって、逆の方向が必要であり、180を加算することでそれを得ます。
3. ベクトル[-2,-4]は、負のxとyを持っています。このベクトルの線は、象限1から象限3へ0,0の点を通ります。-4/-2のTan<sup>-1</sup>は約63.4度で、これはx軸と線との間の角度ですが、やはり逆方向、象限3から象限1へです。正しい方向を反映するために、さらに180度進む必要があります。
4. ベクトル[2,-4]は、正のxと負のyを持っています。これは、象限2から象限4へ0,0の点を通ります。-4/2のTan<sup>-1</sup>は約-64.4度で、これはx軸からベクトル線までの*負の*角度です。技術的には正しいですが、線は-63.4度の角度で右下に移動しています。しかし、*正の*（反時計回り）角度で表したいので、360を加算します。

前のPythonコードでは、数値の正接から逆正接を計算するために*math.**atan***関数を使用しました。*numpy*ライブラリには、同様の***arctan***関数が含まれています。numpy配列を扱う場合、*numpy.**arctan2***関数を使用して、配列ベースのベクトルの逆正接を*ラジアン*で返し、*numpy.**degrees***関数を使用してこれを度数に変換することもできます。***arctan2***関数は、負の*x*と*y*値に対して必要な調整を自動的に行います。

````python
import numpy as np

v = np.array([2,1])
print ('v: ' + str(np.degrees(np.arctan2(v[1], v[0]))))

s = np.array([-3,2])
print ('s: ' + str(np.degrees(np.arctan2(s[1], s[0]))))
````

## ベクトルの加算
これまで、一度に1つのベクトルを扱ってきました。2つのベクトルを加算する必要がある場合はどうなるでしょうか。

例を見てみましょう。すでに、次のように定義された**v**という名前のベクトルがあります。
$$\vec{v} = \begin{bmatrix}2 \\ 1 \end{bmatrix}$$
次に、**s**と呼ばれる2番目のベクトルを次のように作成します。
$$\vec{s} = \begin{bmatrix}-3 \\ 2 \end{bmatrix}$$

以下のセルを実行して、**s**を作成し、**v**と一緒にプロットします。

````python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

v = np.array([2,1])
s = np.array([-3,2])
print (s)

# Plot v and s
vecs = np.array([v,s])
origin = [0], [0]
plt.axis('equal')
plt.grid()
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.quiver(*origin, vecs[:,0], vecs[:,1], color=['r', 'b'], scale=10)
plt.show()
````

プロットを見ると、2つのベクトルが異なる方向と大きさを持っていることがわかります。では、これらを加算するとどうなるでしょうか？

次の式で表されます。
$$\vec{z} = \vec{v}+\vec{s}$$

ベクトル行列で表すと、次のようになります。
$$\vec{z} = \begin{bmatrix}2 \\ 1 \end{bmatrix} + \begin{bmatrix}-3 \\ 2 \end{bmatrix}$$

これにより、次の結果が得られます。
$$\vec{z} = \begin{bmatrix}2 \\ 1 \end{bmatrix} + \begin{bmatrix}-3 \\ 2 \end{bmatrix} = \begin{bmatrix}-1 \\ 3 \end{bmatrix}$$

Pythonで同じ結果が得られることを確認しましょう。

````python
z = v + s
print(z)
````

では、それをプロットで見るとどうなるでしょうか？

````python
vecs = np.array([v,s,z])
origin = [0], [0]
plt.axis('equal')
plt.grid()
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.quiver(*origin, vecs[:,0], vecs[:,1], color=['r', 'b', 'g'], scale=10)
plt.show()
````

何が起こっているのでしょうか？
**s**の次元を**v**の次元に加算して、新しいベクトル**z**を記述しました。これを分解してみましょう。
- **v**の次元は(2,1)なので、開始点から*x*次元に2単位（右に）、*y*次元に1単位（上に）移動します。プロットでは、(0,0)の位置から開始すると、これは赤い矢印として示されます。
- 次に、(-3, 2)の次元を持つ**s**を加算します。したがって、*x*次元に-3単位（負の数なので左に）、*y*次元に2単位（上に）移動します。プロットでは、赤い矢印の頭からこれらの動きをすると、**z**を表す緑の矢印の頭にたどり着きます。

加算操作を逆にして**s**に**v**を加算しても同じです。**s**を作成する手順は青い矢印で記述されており、それを**v**の開始点として使用すると、**z**を表す緑の矢印の頭にたどり着きます。

プロットで、青い矢印の尾を赤い矢印の頭に移動するだけで、その頭が緑の矢印の頭と同じ位置に終わることに注意してください。そして、赤い矢印の尾を青い矢印の頭に移動しても同じことが言えます。
