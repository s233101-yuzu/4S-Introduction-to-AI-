# 変換、固有ベクトル、固有値

行列とベクトルは、空間次元を操作するために一緒に使用されます。これは、3Dコンピューターグラフィックスの数学的生成、幾何学的モデリング、機械学習アルゴリズムのトレーニングと最適化など、多くのアプリケーションで利用されています。ここでは、この主題を網羅的に説明するのではなく、機械学習を扱う際に知っておくと便利ないくつかの重要な概念に焦点を当てます。

## 線形変換
ベクトルを行列で乗算することで、ベクトルを操作できます。行列は、入力ベクトルを操作してベクトル出力を生成する関数として機能します。具体的には、ベクトルの行列乗算は、入力ベクトルを出力ベクトルに変換する*線形変換*です。

たとえば、次の行列***A***とベクトル***v***を考えてみましょう。

$$ A = \begin{bmatrix}2 & 3\\5 & 2\end{bmatrix} \qquad \vec{v} = \begin{bmatrix}1\\2\end{bmatrix}$$

次のように変換***T***を定義できます。

$$ T(\vec{v}) = A\vec{v} $$

この変換を実行するには、*RC*ルールを適用してドット積を計算するだけです。つまり、行列の各行をベクトルの単一の列で乗算します。

$$\begin{bmatrix}2 & 3\\5 & 2\end{bmatrix} \cdot  \begin{bmatrix}1\\2\end{bmatrix} = \begin{bmatrix}8\\9\end{bmatrix}$$

以下は、Pythonでの計算です。

```python
import numpy as np

v = np.array([1,2])
A = np.array([[2,3],
              [5,2]])

t = A@v
print (t)
```

この場合、入力ベクトルと出力ベクトルの両方に2つの成分があります。つまり、変換は2次元ベクトルを受け取り、新しい2次元ベクトルを生成します。これは次のように示されます。

$$ T: \rm I\!R^{2} \to \rm I\!R^{2} $$

出力ベクトルは入力ベクトルとは異なる数の次元を持つ場合があることに注意してください。したがって、行列関数はベクトルをある空間から別の空間に変換する場合があります。または表記法では、${\rm I\!R}$<sup>n</sup> -> ${\rm I\!R}$<sup>m</sup>。

たとえば、ベクトル***v***の元の定義を保持しながら、行列***A***を再定義してみましょう。

$$ A = \begin{bmatrix}2 & 3\\5 & 2\\1 & 1\end{bmatrix} \qquad \vec{v} = \begin{bmatrix}1\\2\end{bmatrix}$$

ここで、もう一度***T***を次のように定義すると、

$$ T(\vec{v}) = A\vec{v} $$

変換は次のように適用されます。

$$\begin{bmatrix}2 & 3\\5 & 2\\1 & 1\end{bmatrix} \cdot  \begin{bmatrix}1\\2\end{bmatrix} = \begin{bmatrix}8\\9\\3\end{bmatrix}$$

したがって、変換はベクトルを2次元空間から3次元空間に変換します。

$$ T: \rm I\!R^{2} \to \rm I\!R^{3} $$

Pythonでは次のとおりです。

```python
import numpy as np
v = np.array([1,2])
A = np.array([[2,3],
              [5,2],
              [1,1]])

t = A@v
print (t)
```

```python
import numpy as np
v = np.array([1,2])
A = np.array([[1,2],
              [2,1]])

t = A@v
print (t)
```

## 大きさと振幅の変換

ベクトルを行列で乗算すると、次の2つのうち少なくとも1つの方法で変換されます。
* 行列の長さ（*大きさ*）をスケーリングして、長くしたり短くしたりする
* 行列の方向（*振幅*）を変更する

たとえば、次の行列とベクトルを考えてみましょう。

$$ A = \begin{bmatrix}2 & 0\\0 & 2\end{bmatrix} \qquad \vec{v} = \begin{bmatrix}1\\0\end{bmatrix}$$

以前と同様に、ベクトル***v***を行列***A***で乗算して変換します。

$$\begin{equation}\begin{bmatrix}2 & 0\\0 & 2\end{bmatrix} \cdot  \begin{bmatrix}1\\0\end{bmatrix} = \begin{bmatrix}2\\0\end{bmatrix}\end{equation}$$

この場合、結果のベクトルは長さ（*大きさ*）が変化しましたが、方向（*振幅*）は変化していません。

Pythonで視覚化してみましょう。

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

v = np.array([1,0])
A = np.array([[2,0],
              [0,2]])

t = A@v
print (t)

# Plot v and t
vecs = np.array([t,v])
origin = [0], [0]
plt.axis('equal')
plt.grid()
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.quiver(*origin, vecs[:,0], vecs[:,1], color=['blue', 'orange'], scale=10)
plt.show()
```

元のベクトル***v***はオレンジ色で、変換されたベクトル***t***は青色で表示されています。***t***は***v***と同じ方向（*振幅*）ですが、長さ（*大きさ*）が大きくなっていることに注意してください。

次に、異なる行列を使用してベクトル***v***を変換してみましょう。
$$\begin{equation}\begin{bmatrix}0 & -1\\1 & 0\end{bmatrix} \cdot  \begin{bmatrix}1\\0\end{bmatrix} = \begin{bmatrix}0\\1\end{bmatrix}\end{equation}$$

今回は、結果のベクトルは異なる振幅に変更されましたが、大きさは同じです。

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

v = np.array([1,0])
A = np.array([[0,-1],
              [1,0]])

t = A@v
print (t)

# Plot v and t
vecs = np.array([v,t])
origin = [0], [0]
plt.axis('equal')
plt.grid()
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.quiver(*origin, vecs[:,0], vecs[:,1], color=['orange', 'blue'], scale=10)
plt.show()
```

もう一度行列を変更してみましょう。
$$\begin{equation}\begin{bmatrix}2 & 1\\1 & 2\end{bmatrix} \cdot  \begin{bmatrix}1\\0\end{bmatrix} = \begin{bmatrix}2\\1\end{bmatrix}\end{equation}$$

これで、結果のベクトルは新しい振幅と大きさの両方に変換されました。変換は方向とスケールの両方に影響を与えました。

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

v = np.array([1,0])
A = np.array([[2,1],
              [1,2]])

t = A@v
print (t)

# Plot v and t
vecs = np.array([v,t])
origin = [0], [0]
plt.axis('equal')
plt.grid()
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.quiver(*origin, vecs[:,0], vecs[:,1], color=['orange', 'blue'], scale=10)
plt.show()
```

### アフィン変換
アフィン変換は、ベクトルを行列で乗算し、オフセットベクトル（*バイアス*とも呼ばれます）を加算します。次のようになります。

$$T(\vec{v}) = A\vec{v} + \vec{b}$$

例：

$$\begin{equation}\begin{bmatrix}5 & 2\\3 & 1\end{bmatrix} \cdot  \begin{bmatrix}1\\1\end{bmatrix} + \begin{bmatrix}-2\\-6\end{bmatrix} = \begin{bmatrix}5\\-2\end{bmatrix}\end{equation}$$

この種の変換は、実際には機械学習の主要な基盤である線形回帰の基礎です。行列は*特徴量*を定義し、最初のベクトルは*係数*、バイアスベクトルは*切片*です。

以下は、Pythonでのアフィン変換の例です。

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

v = np.array([1,1])
A = np.array([[5,2],
              [3,1]])
b = np.array([-2,-6])

t = A@v + b
print (t)

# Plot v and t
vecs = np.array([v,t])
origin = [0], [0]
plt.axis('equal')
plt.grid()
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.quiver(*origin, vecs[:,0], vecs[:,1], color=['orange', 'blue'], scale=15)
plt.show()
```

## 固有ベクトルと固有値

行列を使用してベクトルを変換すると、方向、長さ、またはその両方が変化することがわかりました。変換がスケールのみに影響する場合（つまり、出力ベクトルが入力ベクトルと同じ振幅で異なる大きさを持つ場合）、変換の行列乗算は、ベクトルの何らかのスカラー乗算と同じ操作になります。

たとえば、以前に、ベクトルを行列でドット乗算する次の変換を調べました。

$$\begin{bmatrix}2 & 0\\0 & 2\end{bmatrix} \cdot  \begin{bmatrix}1\\0\end{bmatrix} = \begin{bmatrix}2\\0\end{bmatrix}$$

ベクトルにスカラー値***2***を乗算しても同じ結果が得られます。

$$2 \times \begin{bmatrix}1\\0\end{bmatrix} = \begin{bmatrix}2\\0\end{bmatrix}$$

次のPythonは、これらの計算の両方と結果（同じです）を示しています。

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

v = np.array([1,0])
A = np.array([[2,0],
              [0,2]])

t1 = A@v
print (t1)
t2 = 2*v
print (t2)

fig = plt.figure()
a=fig.add_subplot(1,1,1)
# Plot v and t1
vecs = np.array([t1,v])
origin = [0], [0]
plt.axis('equal')
plt.grid()
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.quiver(*origin, vecs[:,0], vecs[:,1], color=['blue', 'orange'], scale=10)
plt.show()
a=fig.add_subplot(1,2,1)
# Plot v and t2
vecs = np.array([t2,v])
origin = [0], [0]
plt.axis('equal')
plt.grid()
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.quiver(*origin, vecs[:,0], vecs[:,1], color=['blue', 'orange'], scale=10)
plt.show()
```

このような場合、行列変換がスカラーとベクトルの乗算に等しい場合、行列に対応するスカラーとベクトルのペアはそれぞれ固有値と固有ベクトルとして知られています。通常、固有値はギリシャ文字ラムダ（&lambda;）で示され、変換に関する固有値と固有ベクトルを定義する式は次のとおりです。

$$ T(\vec{v}) = \lambda\vec{v}$$

ここで、ベクトル***v***は固有ベクトルであり、値***&lambda;***は変換***T***の固有値です。

変換***T***が行列乗算として表される場合、この場合、変換は行列***A***によって表されます。

$$ T(\vec{v}) = A\vec{v} = \lambda\vec{v}$$

このとき、***v***は行列***A***の固有ベクトルであり、***&lambda;***は固有値です。

行列には複数の固有値-固有ベクトルペアが存在する可能性があり、手動で計算することもできます。ただし、通常はツールやプログラミング言語を使用する方が簡単です。たとえば、Pythonでは、指定された行列の固有値の配列と、対応する固有ベクトルの行列を返す***linalg.eig***関数を使用できます。

以下は、次の行列の固有値と固有ベクトルのペアを返す例です。

$$A=\begin{bmatrix}2 & 0\\0 & 3\end{bmatrix}$$

```python
import numpy as np
A = np.array([[2,0],
              [0,3]])
eVals, eVecs = np.linalg.eig(A)
print(eVals)
print(eVecs)
```

したがって、この行列には2つの固有値-固有ベクトルペアがあり、次のように示されます。

$$ \lambda_{1} = 2, \vec{v_{1}} = \begin{bmatrix}1 \\ 0\end{bmatrix}  \qquad \qquad \lambda_{2} = 3, \vec{v_{2}} = \begin{bmatrix}0 \\ 1\end{bmatrix} $$

各固有値-固有ベクトルペアを乗算すると、固有ベクトルと行列のドット積に対応することを確認してみましょう。最初のペアは次のとおりです。

$$ 2 \times \begin{bmatrix}1 \\ 0\end{bmatrix} = \begin{bmatrix}2 \\ 0\end{bmatrix}  \qquad と \qquad \begin{bmatrix}2 & 0\\0 & 3\end{bmatrix} \cdot \begin{bmatrix}1 \\ 0\end{bmatrix} = \begin{bmatrix}2 \\ 0\end{bmatrix} $$

ここまでは順調です。次に2番目のペアを確認してみましょう。

$$ 3 \times \begin{bmatrix}0 \\ 1\end{bmatrix} = \begin{bmatrix}0 \\ 3\end{bmatrix}  \qquad と \qquad \begin{bmatrix}2 & 0\\0 & 3\end{bmatrix} \cdot \begin{bmatrix}0 \\ 1\end{bmatrix} = \begin{bmatrix}0 \\ 3\end{bmatrix} $$

したがって、固有値-固有ベクトルのスカラー乗算は、行列-固有ベクトルのドット積変換に確かに対応しています。

以下は、前のコードセルで生成した***eVals***変数と***eVecs***変数を使用したPythonでの同等のコードです。

```python
vec1 = eVecs[:,0]
lam1 = eVals[0]

print('Matrix A:')
print(A)
print('-------')

print('lam1: ' + str(lam1))
print ('v1: ' + str(vec1))
print ('Av1: ' + str(A@vec1))
print ('lam1 x v1: ' + str(lam1*vec1))

print('-------')

vec2 = eVecs[:,1]
lam2 = eVals[1]

print('lam2: ' + str(lam2))
print ('v2: ' + str(vec2))
print ('Av2: ' + str(A@vec2))
print ('lam2 x v2: ' + str(lam2*vec2))
```

次のコードを使用して、これらの変換を視覚化できます。

```python
t1 = lam1*vec1
print (t1)
t2 = lam2*vec2
print (t2)

fig = plt.figure()
a=fig.add_subplot(1,1,1)
# Plot v and t1
vecs = np.array([t1,vec1])
origin = [0], [0]
plt.axis('equal')
plt.grid()
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.quiver(*origin, vecs[:,0], vecs[:,1], color=['blue', 'orange'], scale=10)
plt.show()
a=fig.add_subplot(1,2,1)
# Plot v and t2
vecs = np.array([t2,vec2])
origin = [0], [0]
plt.axis('equal')
plt.grid()
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.quiver(*origin, vecs[:,0], vecs[:,1], color=['blue', 'orange'], scale=10)
plt.show()
```

同様に、以前に次の行列変換を調べました。

$$\begin{bmatrix}2 & 0\\0 & 2\end{bmatrix} \cdot  \begin{bmatrix}1\\0\end{bmatrix} = \begin{bmatrix}2\\0\end{bmatrix}$$

そして、ベクトルにスカラー値***2***を乗算しても同じ結果が得られることを確認しました。

$$2 \times \begin{bmatrix}1\\0\end{bmatrix} = \begin{bmatrix}2\\0\end{bmatrix}$$

これは、スカラー値2とベクトル（1,0）がこの行列の固有値-固有ベクトルペアであるために機能します。

Pythonを使用して、この行列の固有値-固有ベクトルペアを決定してみましょう。

```python
import numpy as np
A = np.array([[2,0],
              [0,2]])
eVals, eVecs = np.linalg.eig(A)
print(eVals)
print(eVecs)
```

したがって、この行列にも2つの固有値-固有ベクトルペアがあり、次のように示されます。

$$ \lambda_{1} = 2, \vec{v_{1}} = \begin{bmatrix}1 \\ 0\end{bmatrix}  \qquad \qquad \lambda_{2} = 2, \vec{v_{2}} = \begin{bmatrix}0 \\ 1\end{bmatrix} $$

各固有値-固有ベクトルペアを乗算すると、固有ベクトルと行列のドット積に対応することを確認してみましょう。最初のペアは次のとおりです。

$$ 2 \times \begin{bmatrix}1 \\ 0\end{bmatrix} = \begin{bmatrix}2 \\ 0\end{bmatrix}  \qquad と \qquad \begin{bmatrix}2 & 0\\0 & 2\end{bmatrix} \cdot \begin{bmatrix}1 \\ 0\end{bmatrix} = \begin{bmatrix}2 \\ 0\end{bmatrix} $$

これはすでにわかっていました。次に2番目のペアを確認してみましょう。

$$ 2 \times \begin{bmatrix}0 \\ 1\end{bmatrix} = \begin{bmatrix}0 \\ 2\end{bmatrix}  \qquad と \qquad \begin{bmatrix}2 & 0\\0 & 2\end{bmatrix} \cdot \begin{bmatrix}0 \\ 1\end{bmatrix} = \begin{bmatrix}0 \\ 2\end{bmatrix} $$

次にPythonを使用して、これらの変換を確認し、プロットしてみましょう。

```python
vec1 = eVecs[:,0]
lam1 = eVals[0]

print('Matrix A:')
print(A)
print('-------')

print('lam1: ' + str(lam1))
print ('v1: ' + str(vec1))
print ('Av1: ' + str(A@vec1))
print ('lam1 x v1: ' + str(lam1*vec1))

print('-------')

vec2 = eVecs[:,1]
lam2 = eVals[1]

print('lam2: ' + str(lam2))
print ('v2: ' + str(vec2))
print ('Av2: ' + str(A@vec2))
print ('lam2 x v2: ' + str(lam2*vec2))


# Plot the resulting vectors
t1 = lam1*vec1
t2 = lam2*vec2

fig = plt.figure()
a=fig.add_subplot(1,1,1)
# Plot v and t1
vecs = np.array([t1,vec1])
origin = [0], [0]
plt.axis('equal')
plt.grid()
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.quiver(*origin, vecs[:,0], vecs[:,1], color=['blue', 'orange'], scale=10)
plt.show()
a=fig.add_subplot(1,2,1)
# Plot v and t2
vecs = np.array([t2,vec2])
origin = [0], [0]
plt.axis('equal')
plt.grid()
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.quiver(*origin, vecs[:,0], vecs[:,1], color=['blue', 'orange'], scale=10)
plt.show()
```

## 固有分解

固有値と固有ベクトルについて少し学びました。しかし、それらが何に役立つのか疑問に思っているかもしれません。それらの用途の1つは、変換行列を分解するのに役立つことです。

以前、ベクトルの行列変換がその大きさ、振幅、またはその両方を変化させることを発見しました。技術的な詳細に深く立ち入ることなく、ベクトルが任意の空間方向、つまり*基底*に存在できること、および同じ変換を異なる*基底*に適用できることを覚えておく必要があります。

次の式を使用して行列を分解できます。

$$A = Q \Lambda Q^{-1}$$

ここで、***A***はベクトルをその現在の基底に適用できる変換であり、***Q***は基底の変更を定義する固有ベクトルの行列であり、***&Lambda;***は***Q***によって定義された基底で***A***と同じ線形変換を定義する対角線上に固有値を持つ行列です。

これらを詳しく見てみましょう。次の行列を考えてみましょう。

$$A=\begin{bmatrix}3 & 2\\1 & 0\end{bmatrix}$$

***Q***は、各列が***A***の固有ベクトルである行列です。これは、以前にPythonを使用して計算できることを見ました。

```python
import numpy as np

A = np.array([[3,2],
              [1,0]])

l, Q = np.linalg.eig(A)
print(Q)
```

したがって、行列***A***の場合、***Q***は次の行列です。

$$Q=\begin{bmatrix}0.96276969 & -0.48963374\\0.27032301 & 0.87192821\end{bmatrix}$$

***&Lambda;***は、対角線上に***A***の固有値を持ち、他のすべての要素がゼロである行列です。したがって、2x2行列の場合、&Lambda;は次のようになります。

$$\Lambda=\begin{bmatrix}\lambda_{1} & 0\\0 & \lambda_{2}\end{bmatrix}$$

Pythonコードでは、すでに***linalg.eig***関数を使用して***A***の固有値の配列を***l***変数に返しています。これで、それを行列としてフォーマットするだけです。

```python
L = np.diag(l)
print (L)
```

したがって、***&Lambda;***は次の行列です。

$$\Lambda=\begin{bmatrix}3.56155281 & 0\\0 & -0.56155281\end{bmatrix}$$

次に、***Q***の逆行列である***Q<sup>-1</sup>***を見つけるだけです。

```python
Qinv = np.linalg.inv(Q)
print(Qinv)
```

***Q***の逆行列は次のとおりです。

$$Q^{-1}=\begin{bmatrix}0.89720673 & 0.50382896\\-0.27816009 & 0.99068183\end{bmatrix}$$

それは何を意味するのでしょうか？つまり、行列***A***を乗算した*任意の*ベクトルの変換を、個別の操作***Q&Lambda;Q<sup>-1</sup>***に分解できるということです。

$$A\vec{v} = Q \Lambda Q^{-1}\vec{v}$$

これを証明するために、ベクトル***v***をとってみましょう。

$$\vec{v} = \begin{bmatrix}1\\3\end{bmatrix} $$

行列***A***を使用した行列変換は次のとおりです。

$$\begin{bmatrix}3 & 2\\1 & 0\end{bmatrix} \cdot \begin{bmatrix}1\\3\end{bmatrix} $$

それでは、Pythonを使用してその結果を表示しましょう。

```python
v = np.array([1,3])
t = A@v

print(t)

# Plot v and t
vecs = np.array([v,t])
origin = [0], [0]
plt.axis('equal')
plt.grid()
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.quiver(*origin, vecs[:,0], vecs[:,1], color=['orange', 'b'], scale=20)
plt.show()
```

次に、***Q&Lambda;Q<sup>-1</sup>***の操作シーケンスを使用して同じことを行います。

```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

t = (Q@(L@(Qinv)))@v

# Plot v and t
vecs = np.array([v,t])
origin = [0], [0]
plt.axis('equal')
plt.grid()
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.quiver(*origin, vecs[:,0], vecs[:,1], color=['orange', 'b'], scale=20)
plt.show()
```

したがって、***A***と***Q&Lambda;Q<sup>-1</sup>***は同等です。

分解された変換の中間段階を見ると、***v***の元の基底における***A***を使用した変換（オレンジから青）、および***Q***によって記述された基底の変更における***&Lambda;***を使用した変換（赤からマゼンタ）を確認できます。

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

t1 = Qinv@v
t2 = L@t1
t3 = Q@t2

# Plot the transformations
vecs = np.array([v,t1, t2, t3])
origin = [0], [0]
plt.axis('equal')
plt.grid()
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.quiver(*origin, vecs[:,0], vecs[:,1], color=['orange', 'red', 'magenta', 'blue'], scale=20)
plt.show()
```

この視覚化から、変換***Av***は、***Q***を使用して***v***の基底を変更し（上記のプロットではオレンジから赤）、***&Lambda;***を使用してその基底で同等の線形変換を適用し（赤からマゼンタ）、***Q<sup>-1</sup>***を使用して元の基底に戻る（マゼンタから青）ことによって実行できることが明らかになるはずです。

## 行列のランク

正方行列の**ランク**は、その行列の非ゼロ固有値の数です。**フルランク**の行列は、行列の次元と同じ数の非ゼロ固有値を持っています。**ランク落ち**の行列は、次元よりも少ない非ゼロ固有値を持っています。ランク落ちの行列の逆行列は特異であり、存在しません（これが以前のノートブックで一部の行列には逆行列がないと述べた理由です）。

次の行列***A***を考えてみましょう。

$$A=\begin{bmatrix}1 & 2\\4 & 3\end{bmatrix}$$

その固有値（***&Lambda;***）を見つけてみましょう。

```python
import numpy as np
A = np.array([[1,2],
              [4,3]])
l, Q = np.linalg.eig(A)
L = np.diag(l)
print(L)
```

$$\Lambda=\begin{bmatrix}-1 & 0\\0 & 5\end{bmatrix}$$

この行列はフルランクです。行列の次元は2です。非ゼロ固有値は2つあります。

次に、この行列を考えてみましょう。

$$B=\begin{bmatrix}3 & -3 & 6\\2 & -2 & 4\\1 & -1 & 2\end{bmatrix}$$

2列目と3列目が1列目のスカラー倍であることに注意してください。

その固有値を調べてみましょう。

```python
B = np.array([[3,-3,6],
              [2,-2,4],
              [1,-1,2]])
lb, Qb = np.linalg.eig(B)
Lb = np.diag(lb)
print(Lb)
```

$$\Lambda=\begin{bmatrix}3 & 0& 0\\0 & -6\times10^{-17} & 0\\0 & 0 & 3.6\times10^{-16}\end{bmatrix}$$

この行列には非ゼロ固有値が1つしかないことに注意してください。他の2つの固有値は非常に小さく、実質的にゼロです。これはランク落ち行列の例であり、そのため逆行列を持ちません。

## 正方フルランク行列の逆行列

正方フルランク行列の逆行列は、次の式を使用して計算できます。

$$A^{-1} = Q \Lambda^{-1} Q^{-1}$$

これを行列***A***に適用してみましょう。

$$A=\begin{bmatrix}1 & 2\\4 & 3\end{bmatrix}$$

***Q***、***&Lambda;<sup>-1</sup>***、および***Q<sup>-1</sup>***の行列を見つけてみましょう。

```python
import numpy as np
A = np.array([[1,2],
              [4,3]])

l, Q = np.linalg.eig(A)
L = np.diag(l)
print(Q)
Linv = np.linalg.inv(L)
Qinv = np.linalg.inv(Q)
print(Linv)
print(Qinv)
```

したがって、

$$A^{-1}=\begin{bmatrix}-0.70710678 & -0.4472136\\0.70710678 & -0.89442719\end{bmatrix}\cdot\begin{bmatrix}-1 & -0\\0 & 0.2\end{bmatrix}\cdot\begin{bmatrix}-0.94280904 & 0.47140452\\-0.74535599 & -0.74535599\end{bmatrix}$$

Pythonでこれを計算してみましょう。

```python
Ainv = (Q@(Linv@(Qinv)))
print(Ainv)
```

これにより、次の結果が得られます。

$$A^{-1}=\begin{bmatrix}-0.6 & 0.4\\0.8 & -0.2\end{bmatrix}$$

これを検証するために、***np.linalg.inv***関数を***A***に直接適用できます。

```python
print(np.linalg.inv(A))
```
